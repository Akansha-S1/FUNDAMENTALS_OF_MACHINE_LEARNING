## Explore and visualize the missing data patterns. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_file_path = r'C:\Users\akans\OneDrive\Desktop\SEM_3\TERM_7\ML_T7\HousingData.csv'

df = pd.read_csv(csv_file_path, na_values=['N/A'])

missing_data = df.isnull()

plt.figure(figsize=(14, 8))
sns.heatmap(missing_data, cmap='YlGnBu', cbar=True, linewidths=0.5, linecolor='white', annot=False)
plt.title('Missing Data Patterns')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()

plt.figure(figsize=(12, 6))
missing_percentage = (missing_data.sum() / len(df)) * 100
missing_percentage.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Percentage of Missing Values in Each Column')
plt.xlabel('Columns')
plt.ylabel('Percentage')
plt.show()
```

## Handle missing values using imputation method for a specific feature. 

```python
import pandas as pd
from sklearn.impute import SimpleImputer

csv_file_path = r'C:\Users\akans\OneDrive\Desktop\SEM_3\TERM_7\ML_T7\HousingData.csv'

df = pd.read_csv(csv_file_path, na_values=['N/A'])

specific_feature = 'CRIM'

imputer = SimpleImputer(strategy='mean')

df[specific_feature] = imputer.fit_transform(df[[specific_feature]])
```

```python
missing_values_crim = df['CRIM'].isnull().sum()

if missing_values_crim == 0:
    print("The 'CRIM' column has no missing values after imputation.")
else:
    print(f"The 'CRIM' column still has {missing_values_crim} missing values.")
```
